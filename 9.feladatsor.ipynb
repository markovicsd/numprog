#### 0. Feladat:

Ismerjük meg a különböző [ábrázolási lehetőségeket](https://ch.mathworks.com/help/matlab/creating_plots/types-of-matlab-plots.html). Fordítsunk különös figyelmet a `plot`, `loglog`, `histogram`, `scatter`, `contour`, `countourf`, `surf` parancsokra. 

```octave
x = linspace(0, 1, 4)
[X,Y] = meshgrid(x,x)
```

```octave
x = linspace(0,1, 100);
y = linspace(0,1, 100);
[X,Y] = meshgrid(x,y);
```

```octave
contourf(X, Y, X.^2+Y.^2)
```

```octave
surf(X, Y, X.^2+Y.^2)
```
